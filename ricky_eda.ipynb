```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
file_path_1 = 'Netflix_cleaned_1.csv'
file_path_2 = 'Hulu_cleaned_1.csv'
df_netflix = pd.read_csv(file_path_1)
df_hulu = pd.read_csv(file_path_2)
```

```python
genre_counts = {}
for row in df_hulu.iterrows():
    lst = row[1]['listed_in'].split(',')
    for genre in lst:
        genre = genre.strip()
        if genre not in genre_counts:
            genre_counts[genre] = 1
        else:
            genre_counts[genre] += 1
```

```python
genre_counts = {k: v for k, v in sorted(genre_counts.items(), key=lambda item: item[1], reverse=True)}
genre_counts
```

```python
genre_counts = {}
for row in df_netflix.iterrows():
    lst = row[1]['listed_in'].split(',')
    for genre in lst:
        genre = genre.strip()
        if genre not in genre_counts:
            genre_counts[genre] = 1
        else:
            genre_counts[genre] += 1
```

```python
genre_counts = {k: v for k, v in sorted(genre_counts.items(), key=lambda item: item[1], reverse=True)}
genre_counts
```

```python
def one_hot_genre(listed_in, genre):
    lst = list(set(listed_in.split(',')))
    for i in lst:
        if genre in i:
            return 1
    return 0
```

```python
hulu_genres_df = df_hulu[['type', 'month', 'year', 'continent', 'listed_in']].copy()
net_genres_df = df_netflix[['type', 'month', 'year', 'continent', 'listed_in']].copy()
genre_list = ['international', 'drama', 'comed', 'action', 'docu', 'thriller', 'roman']
for genre in genre_list:
    hulu_genres_df[genre] = hulu_genres_df['listed_in'].apply(lambda row: one_hot_genre(row, genre))
    net_genres_df[genre] = net_genres_df['listed_in'].apply(lambda row: one_hot_genre(row, genre))

hulu_genres_df['other'] = hulu_genres_df[genre_list].apply(lambda row: 1 if (row == 0).all() else 0, axis=1)
net_genres_df['other'] = net_genres_df[genre_list].apply(lambda row: 1 if (row == 0).all() else 0, axis=1)
```

```python
def get_genre(genre):
    if genre=='international':
        return 'International'
    elif genre=='drama':
        return 'Drama'
    elif genre=='comed':
        return 'Comedy'
    elif genre=='action':
        return 'Action'
    elif genre=='docu':
        return 'Documentary'
    elif genre=='thriller':
        return 'Thriller'
    elif genre=='roman':
        return 'Romance'
    elif genre=='other':
        return 'Other'
```

```python
fig, ax = plt.subplots(figsize=(20,8), nrows=4, ncols=2)
ax_ind = [[0,0], [0,1], [1,0], [1,1], [2,0], [2,1], [3,0], [3,1]]
genre_list = ['international', 'drama', 'comed', 'action', 'docu', 'thriller', 'roman', 'other']
for ind, genre in zip(ax_ind, genre_list):
    hulu_grouped = hulu_genres_df.groupby('year')[genre].mean()
    net_grouped = net_genres_df.groupby('year')[genre].mean()
    ax[ind[0], ind[1]].plot(hulu_grouped, color = '#1ce783', label='Hulu')
    ax[ind[0], ind[1]].plot(net_grouped, color = '#E50914', label='Netflix')
    ax[ind[0], ind[1]].set_title(get_genre(genre))
    ax[ind[0], ind[1]].spines['top'].set_visible(False)
    ax[ind[0], ind[1]].spines['right'].set_visible(False)
    ax[ind[0], ind[1]].set_facecolor(color='#f7f4ed')

handles, labels = ax[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.47, 0.925), ncol=2, fontsize=14)
fig.suptitle('Proportion of Movies and Shows Released by Genre Over Time', weight='bold', x=0.47, ha='center', size=16)
fig.supxlabel('Year Added', x=0.47, ha='center', size=14)
fig.supylabel('Percent of Movies and Shows Added', size=14)
fig.patch.set_facecolor('#f7f4ed')
plt.subplots_adjust(wspace=0.1, hspace=0.5)
plt.subplots_adjust(top=0.85)
plt.subplots_adjust(left=0.0555)
plt.show()
    
```

```python
colormap = {
    'international': '#FF5733',
    'drama': '#33FF57',
    'comed': '#3357FF',
    'action': '#FF33A5',
    'docu': '#A533FF',
    'thriller': '#33FFA5',
    'roman': '#FFA533',
    'other': '#5733FF'
}
```

```python
fig, ax = plt.subplots(figsize=(20,8), nrows=1, ncols=2)
genre_list = ['international', 'drama', 'comed', 'action', 'docu', 'thriller', 'roman']
for genre in genre_list:
    hulu_grouped = hulu_genres_df.groupby('year')[genre].sum()
    net_grouped = net_genres_df.groupby('year')[genre].sum()
    ax[0].plot(hulu_grouped, color = colormap[genre], label=get_genre(genre))
    ax[1].plot(net_grouped, color = colormap[genre], label=get_genre(genre))

ax[0].set_title('Hulu', fontsize=16, weight='bold')
ax[1].set_title('Netflix', fontsize=16, weight='bold')
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[0].set_facecolor(color='#f7f4ed')
ax[1].set_facecolor(color='#f7f4ed')

handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right', bbox_to_anchor=(0.9, 0.975), ncol=2, fontsize=10)
fig.suptitle('Total Shows and Movies Added By Genre Over Time', weight='bold', x=0.47, ha='center', size=18)
fig.supxlabel('Year Added', x=0.47, ha='center', size=14)
fig.supylabel('# of Movies and Shows Added', size=14)
fig.patch.set_facecolor('#f7f4ed')
plt.subplots_adjust(wspace=0.1, hspace=0.5)
plt.subplots_adjust(top=0.85)
plt.subplots_adjust(left=0.0555)
plt.show()
```
